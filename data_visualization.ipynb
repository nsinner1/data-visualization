```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
# Wine Reviews
```

```python
wine = pd.read_csv("./wine.csv", error_bad_lines=False)
wine.head()
```

```python
type(wine)
```

```python
# Does the point rating determine price of wine?
```

```python
sns.relplot(x="price", y="points", data=wine)
```

```python
# Does country determine price?
```

```python
sns.relplot(x="country", y="price", data=wine, hue="country")
```

```python
# Does variety determine price?
```

```python
sns.barplot(x="variety", y="price", data=wine)
```
